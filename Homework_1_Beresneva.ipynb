- Data set is "Meta-data Data Set" https://archive.ics.uci.edu/ml/datasets/Meta-data
- Relevant Information: 
> This DataSet is about the results of Statlog project. The project performed a comparative study between Statistical, Neural and Symbolic learning algorithms. 
- Number of Instances: 528
- Number of Attributes: 22
- Attribute Information:
 - `DS_Name` - Name of DataSet 
 - `T` - Number of examples in test set
 - `N` - Number of examples
 - `p` - Number of attributes
 - `k` - Number of classes
 - `Bin` - Number of binary Attributes 
 - `Cost` - Cost ( `1 = yes`,`0 = no`)
 - `SDratio` - Standard deviation ratio 
 - `correl`- Mean correlation between attributes
 - `cancor1` - First canonical correlation
 - `cancor2` - Second canonical correlation
 - `fract1` - First eigenvalue 
 - `fract2` - Second eigenvalue 
 - `skewness` - Mean of $ \frac{|E(x - Mean)|^3}{STD^3} $
 - `kurtosis` - Mean of $ \frac{|E(x - Mean)|^4}{STD^4} $
 - `Hc` - Mean entropy of attributes
 - `Hx`- Entropy of classes - энтропия классов
 - `MCx` - Mean mutual entropy of class and attributes
 - `EnAtr` - Equivalent number of attributes - Эквивалентное количество атрибутов
 - `NSRatio` - Noise-signal ratio - Отношение сигнал / шум
 - `Alg_Name` - Name of Algorithm 
 - `Norm_error` - Normalized Error (continuous class)

```python
y_train = np.array([1 if it == 1 else 0 for it in y_train])
y_test = np.array([1 if it == 1 else 0 for it in y_test])
```

# $k$NN

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 10)
#knn.set_params(n_neighbors=10)
knn.fit(X_train, y_train)
```

```python
knn
```

```python
y_train_predict = knn.predict(X_train)
y_test_predict = knn.predict(X_test)

err_train = np.mean(y_train != y_train_predict)
err_test  = np.mean(y_test  != y_test_predict)

print(err_train, err_test)
```

Значения ошибок на обучающей и тестовой выборках

```python
err_test = 1 - knn.score(X_test, y_test)
err_test
```

Другой способ нахождения ошибки

```python
from sklearn.metrics import confusion_matrix

print(confusion_matrix(y_test, y_test_predict))
```

# Подбор параметров

```python
from sklearn.model_selection import GridSearchCV
nnb = [1, 3, 5, 10, 15, 20, 25, 35, 45, 55]
knn = KNeighborsClassifier()
grid = GridSearchCV(knn, param_grid = {'n_neighbors': nnb}, cv=10)
grid.fit(X_train, y_train)

best_cv_err = 1 - grid.best_score_
best_n_neighbors = grid.best_estimator_.n_neighbors
print(best_cv_err, best_n_neighbors)
```

Метод подобрал $k=3$

```python
knn = KNeighborsClassifier(n_neighbors = best_n_neighbors).fit(X_train, y_train)

err_train = np.mean(y_train != knn.predict(X_train))
err_test  = np.mean(y_test  != knn.predict(X_test))

print(err_train, err_test)
```
